$(x_1, \dots, x_n) \sim U_{[0,\theta]} -$ выборка. Оцненка параметра имеет виду $\theta^* = 2 \overline{X}$

Необходимо продемонстрировать несмещенность и исследовать корость сходимости к значению параметра.

Эта оценка дейтвительно несмещена, так как
$$E \theta^* = E 2 \overline{X} = 2* \theta/2 = \theta$$

```python
import numpy as np
from matplotlib import pyplot as plt

theta = 5

vector = np.random.uniform(0,theta,1)
means = []
for length in np.arange(10,500,1):
    preds = []
    for sample in range(100):
        vector = np.random.uniform(0,theta,length)
        theta_pred = 2* np.mean(vector)
        preds.append(theta_pred)
    mean = np.mean(preds)
    means.append(mean)
plt.plot((means))
```

Видно, что этот график раскладывается на 5 + var, где var -- случайная величина у которой уменьшается дисперис с такой скоростью, какая скорость сходимости у оценки.

```python
import numpy as np
from matplotlib import pyplot as plt

theta = 5

vector = np.random.uniform(0,theta,1)
vars = []
for length in np.arange(2,5000,1):
    preds = []
    for sample in range(100):
        vector = np.random.uniform(0,theta,length)
        theta_pred = 2* np.mean(vector)
        preds.append(abs(theta_pred - theta))
    var = np.mean(preds) 
    vars.append(abs(var))
plt.plot((vars))
```

```python
li = []
for base in np.arange(1,5000 // 2 -1, 5):
    li.append(vars[base]/vars[base*2])
```

```python
plt.plot(li)
```

```python
np.mean(li)
```

Видно, что с увеличением выборки в 2 раза, предсказание отклонение улучшается в 1.4 раза, то есть скорость сходимости имеет порядок 1/sqrt(n)
